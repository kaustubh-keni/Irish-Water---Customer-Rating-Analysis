```python
# Columns to be dropped

columns_to_drop = ['responseid', 'Invite Date', 'Callback request',
       'CaseID', 'Case type', 'Resolution Code', 'Reasons for resolution outcome',
       'Reasons for overall satisfaction score',
       'Single word', 'Suggestions for improvement', 'Listened to and understood',
       'Next steps were explained', 'Kept informed', 'Resolution time frame',
       'How could contact centre deal better?']

# Drop specified columns

df.drop(columns = columns_to_drop, inplace = True)

#Verify columns dropped

df.head(5)
```

```python
# Duplicate the column Resolution

df['Resolution Feedback'] = df['Resolution']

# Verify new duplicate column added

df.head(5)
```

```python
# Replace values to simplify

# Defining the replacements

replacements_j = { 'Operations Fieldwork' : 'Operations',
                   'Commercial - Billing Query' : 'Commercial'
                 }


replacements_cc = { 'Carry out part replacement repair of <=32mm MDPE potable water service pipe' : 'Part replacement repair',
                    'Waste Action - Unblock Sewer - Below Ground Waste Water' : 'Unblock Sewer',
                    'Investigation No Water - Below Ground Drinking Water' : 'No Water',
                    'Investigation Blockage - Below Ground Waste Water' : 'Blockage',
                    'Repair Main - Below Ground Drinking Water' : 'Repair Mains',
                    'Investigation Visible Leak - Below Ground Drinking Water' : 'Visible Leak'
                  }


replacements_cq = { 'You recently contacted Uisce Ã‰ireann/Irish Water to change something on your account' : 'Change of Account Details',
                   'You recently contacted Uisce Ã‰ireann/Irish Water to make a card payment' : 'Card Payment',
                   'You recently contacted Uisce Ã‰ireann/Irish Water with a query on a blockage' : 'Blockage',
                   'You recently contacted Uisce Ã‰ireann/Irish Water with a query on a leakage' : 'Leakage',
                   'You recently contacted Uisce Ã‰ireann/Irish Water with a query on no water' : 'No Water',
                   'You recently contacted Uisce Ã‰ireann/Irish Water with a query on payment' : 'Payment Query',
                   'You recently contacted Uisce Ã‰ireann/Irish Water with an account balance request' : 'Account Balance Request'
                 }


replacements_r = { "YES, it was ^f(\"Intropipe\").any('4') ? \"completed\": \"resolved\"^ to my satisfaction": 'Yes',
                   "YES, it was ^f(\"Intropipe\").any('4') ? \"completed\": \"resolved\"^ but not to my satisfaction": 'Yes',
                   "NO, it has still not been ^f(\"Intropipe\").any('4') ? \"completed\": \"resolved\"^/still ongoing": 'No'
                 }


replacements_rf = { "YES, it was ^f(\"Intropipe\").any('4') ? \"completed\": \"resolved\"^ to my satisfaction": 'Satisfied',
                    "YES, it was ^f(\"Intropipe\").any('4') ? \"completed\": \"resolved\"^ but not to my satisfaction": 'Not Satisfied',
                    "NO, it has still not been ^f(\"Intropipe\").any('4') ? \"completed\": \"resolved\"^/still ongoing": 'Still Ongoing'
                  }


replacements_os = { '10 - Extremely Satisfied' : 10,
                   '1 - Extremely Dissatisfied' : 1
                 }


replacements_e = { '7 - Extremely Easy' : 7,
                  '1 - Extremely Difficult' : 1
                }


# Replace in the Case Characteristic column

df['Journey'] = df['Journey'].replace(replacements_j)

df['Case Characteristic'] = df['Case Characteristic'].replace(replacements_cc)

df['Customer Query'] = df['Customer Query'].replace(replacements_cq)

df['Resolution'] = df['Resolution'].replace(replacements_r)

df['Resolution Feedback'] = df['Resolution Feedback'].replace(replacements_rf)

df['Overall satisfaction'] = df['Overall satisfaction'].replace(replacements_os)

df['Easy'] = df['Easy'].replace(replacements_e)

df['Satisfaction with contact centre experience'] = df['Satisfaction with contact centre experience'].replace(replacements_os)

df['Next steps clearly explained'] = df['Next steps clearly explained'].replace(replacements_os)

df['Query resolution'] = df['Query resolution'].replace(replacements_os)

df['Satisfaction with Engineer'] = df['Satisfaction with Engineer'].replace(replacements_os)

df["Agents' ability"] = df["Agents' ability"].replace(replacements_os)

df['Resolution time for query'] = df['Resolution time for query'].replace(replacements_os)


# Verify the replacements have been made

print(df[['Case Characteristic', 'Customer Query',  'Resolution', 'Resolution Feedback', 'Overall satisfaction',
          'Easy','Journey', 'Satisfaction with contact centre experience', 'Next steps clearly explained', 
          'Query resolution', 'Satisfaction with Engineer', "Agents' ability", 'Resolution time for query']].head(5))
```

```python
# Dropping few more columns, Columns to be dropped

additional_columns_to_drop = ['PersonID', 'WorkorderID', 'CC_ID', 'SRNumber' ]

# Drop specified columns

df.drop(columns = additional_columns_to_drop, inplace = True)

#Verify columns dropped

df.head(5)
```

```python
# Looking at the null values 

df.isnull().sum()
```

```python
# Convert required datatypes

df['Satisfaction with contact centre experience'] = df['Satisfaction with contact centre experience'].astype(float)

df['Next steps clearly explained'] = df['Next steps clearly explained'].astype(float)

df['Query resolution'] = df['Query resolution'].astype(float)

df['Satisfaction with Engineer'] = df['Satisfaction with Engineer'].astype(float)

df["Agents' ability"] = df["Agents' ability"].astype(float)

df['Resolution time for query'] = df['Resolution time for query'].astype(float)

# Adjust rating for 'Easy' column from 1 to 10 for easier comparision
# Convert 'Easy' column to float to ensure division works correctly
df['Easy'] = df['Easy'].astype(float)

# Scale the 'Easy' column from range [1, 7] to [1, 10]
df['Easy'] = ((df['Easy'] - 1) / (7 - 1)) * (10 - 1) + 1

# Convert the scaled values to integers (if desired)
df['Easy'] = df['Easy'].round().astype(int)

# Verify the changes

df.head(5)
```
